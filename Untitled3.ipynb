```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("quikr_car.csv")

df.head()
```

```python
print("Shape of Dataset:")
print(df.shape)

print("\nColumns:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Remove Ask For Price rows
df = df[df["Price"] != "Ask For Price"]

# Clean Price
df["Price"] = (
    df["Price"]
    .astype(str)
    .str.replace(",", "")
)

df = df[df["Price"].str.isnumeric()]
df["Price"] = df["Price"].astype(int)

# Clean kms_driven
df["kms_driven"] = (
    df["kms_driven"]
    .astype(str)
    .str.replace(" kms", "")
    .str.replace(",", "")
)

df = df[df["kms_driven"].str.isnumeric()]
df["kms_driven"] = df["kms_driven"].astype(int)

# Clean Year
df = df[df["year"].astype(str).str.isnumeric()]
df["year"] = df["year"].astype(int)

# Remove missing values
df.dropna(inplace=True)

df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,6))

df["company"].value_counts().head(10).plot(
    kind="bar"
)

plt.title("Top 10 Car Brands")
plt.xlabel("Brand")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(7,7))

df["fuel_type"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Fuel Type Distribution")

plt.ylabel("")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    df["Price"],
    bins=30,
    kde=True
)

plt.title("Price Distribution")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    df["kms_driven"],
    bins=30,
    kde=True
)

plt.title("Kilometers Driven Distribution")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df,
    x="kms_driven",
    y="Price"
)

plt.title("Price vs Kilometers Driven")

plt.show()
```

```python
avg_price = (
    df.groupby("company")["Price"]
      .mean()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(10,6))

avg_price.plot(kind="bar")

plt.title("Top 10 Brands by Average Price")

plt.show()
```

```python
numeric_df = df[
    ["Price",
     "year",
     "kms_driven"]
]

corr = numeric_df.corr()

corr
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
print("EDA INSIGHTS")
print("="*40)

print("\n1. Most Popular Brand:")
print(df["company"].value_counts().idxmax())

print("\n2. Most Common Fuel Type:")
print(df["fuel_type"].value_counts().idxmax())

print("\n3. Average Car Price:")
print(round(df["Price"].mean(),2))

print("\n4. Average Kilometers Driven:")
print(round(df["kms_driven"].mean(),2))

print("\n5. Newest Car Year:")
print(df["year"].max())
```
